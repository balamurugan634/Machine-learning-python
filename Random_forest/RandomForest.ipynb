```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('cardekho_data.csv')
```

```python
df
```

```python
for_train=df.drop(['Car_Name'],axis=1)
```

```python
for_train
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder=LabelEncoder()
df['enc_transmission']=encoder.fit_transform(df['Transmission'])
```

```python
df.enc_transmission
```

```python
df
```

```python
df['enc_Fuel_Type']=encoder.fit_transform(df['Fuel_Type'])
df['enc_Seller_Type']=encoder.fit_transform(df['Seller_Type'])
```

```python
df
```

```python
for_train=df.drop(['Car_Name','Fuel_Type','Seller_Type','Transmission','Owner'],axis=1)
y=np.array(df['Owner'])
```

```python
y
```

```python
for_train
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(for_train,y,test_size=0.2)
```

```python
X_train
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(n_estimators=20,random_state=30)
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```
