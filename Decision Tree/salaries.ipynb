```python
enc=LabelEncoder()
inputs['comp_n']=enc.fit_transform(inputs.company)
inputs['job_n']=enc.fit_transform(inputs.job)
inputs['deg_n']=enc.fit_transform(inputs.degree)
```

```python
inputs
```

```python
inputs=inputs.drop(['company','job','degree'],axis=1)
```

```python
inputs
```

```python
X_train,X_test,y_train,y_test=train_test_split(inputs,op,test_size=0.2)
```

```python
model=tree.DecisionTreeClassifier()
model.fit(X_train,y_train)
```
